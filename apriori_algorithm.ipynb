# Loading dataset

```python
import pandas as pd
```

```python
data=pd.read_csv("basket.csv")
data
```

```python
data.shape
```

# Replacing columns name

```python
data.columns = [f'col_{j}' for j in range(1, 12)]
print(data.columns)
```

```python
data.head()
```

# creating list

```python
market=[]
for i in range(0,data.shape[0]):
    cus=[]
    for j in data.columns:
        if type(data[j][i])==str:
            cus.append(data[j][i])
    market.append(cus)
```

```python
market[0:7]
```

```python
l = []

for i in market:
    for j in i:
        l.append(j)
        
```

# Making dictionary

```python
import collections
```

```python
p=collections.Counter(l)
```

```python
d={'item_name':p.keys(),'values':p.values()}
pd.DataFrame(d).sort_values(by='values',ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
tr=TransactionEncoder()
tr.fit(market)
```

```python
tr.transform(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
df
```

# apriori

```python
from mlxtend.frequent_patterns import apriori
apriori(df,min_support=0.07,use_colnames=True,max_len=3).sort_values(by='support')
```
